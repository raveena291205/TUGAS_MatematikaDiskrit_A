```python
import matplotlib.pyplot as plt

# Mengecek nilai n berapa saja yang membuat
# 1 + 1/2 + 1/3 + ... + 1/n < 2
hasil = []

for n in range(1, 20):  
    jumlah = sum(1/i for i in range(1, n+1))
    hasil.append((n, jumlah, jumlah < 2))

# Cetak hasilnya
print("n\tJumlah\t\t< 2?")
for n, jumlah, valid in hasil:
    print(f"{n}\t{jumlah:.6f}\t{valid}")

# --- Buat Grafik ---
n_values = [h[0] for h in hasil]
jumlah_values = [h[1] for h in hasil]

plt.figure(figsize=(8,5))
plt.plot(n_values, jumlah_values, marker='o', label="Jumlah 1 + 1/2 + ... + 1/n")
plt.axhline(y=2, color='r', linestyle='--', label="Batas = 2")  # garis horizontal di y=2

plt.title("Grafik Deret Harmonik Parsial")
plt.xlabel("n")
plt.ylabel("Jumlah")
plt.legend()
plt.grid(True)
plt.show()
```

Interpretasi Hasil Deret Harmonik Parsial

1. Deret harmonik adalah penjumlahan pecahan 1 + 1/2 + 1/3 + ... + 1/n , Grafik menunjukkan bahwa nilai deret ini selalu bertambah seiring bertambahnya 

2. Dari tabel dan grafik terlihat:
   Saat n = 1, jumlahnya = 1.0000 (<2)
   Saat n = 2, jumlahnya = 1.5000 (<2)
   saat n = 3, jumlahnya = 1.8333 (<2)
   saat n = 4, jumlahnya = 2.0833 (sudah lebih besar dari 2)
   Jadi, nilai terbesar n yang masih memenuhi kondisi jumlah < 2 adalah n = 3

3. Grafik menunjukkan garis biru yang naik terus (hasil penjumlahan deret harmonik), sedangkan garis merah putus-putus menunjukkan batas angka 2.
   Titik-titik sampai n = 3 berada dibawah garis batas
   mulai n = 4, titik-titik sudah menembus garis batas artinya jumlah deret sudah melampaui 2

4. Dari sini bisa disimpulkan bahwa deret harmonik tumbuh sangat lambat (bertambah kecil setiap pecahannya), tetapi pada akhirnya akan terus naik tanpa batas (divergen). Dalam kasus ini, batas pertama kali dilampaui ketika n=4
