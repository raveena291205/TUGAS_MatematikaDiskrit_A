```python
def verify_statement_1(n):
    """Verifikasi: 1 + 3 + 5 + ... + (2n - 1) = n²"""
    if n <= 0:
        return False, "n harus > 0"
    
    # Hitung sisi kiri: jumlah n bilangan ganjil pertama
    left_side = sum(2*i - 1 for i in range(1, n + 1))
    
    # Sisi kanan: n²
    right_side = n**2
    
    return left_side == right_side, f"{left_side} {'==' if left_side == right_side else '!='} {right_side}"

def verify_statement_5(n):
    """Verifikasi: 2ⁿ > n²"""
    if n < 0:
        return False, "n harus >= 0"
    
    # Hitung 2^n
    power_2 = 2**n
    
    # Hitung n²
    n_squared = n**2
    
    return power_2 > n_squared, f"{power_2} {'>' if power_2 > n_squared else '<='} {n_squared}"

def main():
    print("=" * 60)
    print("VERIFIKASI PERNYATAAN MATEMATIKA")
    print("=" * 60)
    
    statements = [
        ("1 + 3 + 5 + ... + (2n - 1) = n²", verify_statement_1),
        ("1 + 2 + 2² + 2³ + ... + 2ⁿ = 2^(n+1) - 1", verify_statement_2),
        ("1² + 2² + 3² + ... + n² = n(n+1)(2n+1)/6", verify_statement_3),
        ("n! >= 2^(n-1)", verify_statement_4),
        ("2ⁿ > n²", verify_statement_5)
    ]
    
    # Test range yang lebih fokus
    test_range = range(1, 15)  # Fokus pada range yang lebih relevan
    
    for i, (description, func) in enumerate(statements, 1):
        print(f"\n{i}. {description}")
        print("-" * 50)
        
        valid_values = []
        invalid_values = []
        
        for n in test_range:
            try:
                is_valid, detail = func(n)
                if is_valid:
                    valid_values.append(n)
                    print(f"n = {n:2d}: ✓ {detail}")
                else:
                    invalid_values.append(n)
                    if len(invalid_values) <= 5:  # Hanya tampilkan beberapa yang invalid
                        print(f"n = {n:2d}: ✗ {detail}")
            except Exception as e:
                print(f"n = {n:2d}: ERROR - {e}")
        
        # Analisis pola
        print(f"\nHasil Analisis:")
        if valid_values:
            if len(valid_values) == len(test_range):
                print(f"✓ Valid untuk semua n dalam range {min(test_range)}-{max(test_range)}")
            else:
                # Cari pola
                consecutive_groups = []
                current_group = [valid_values[0]] if valid_values else []
                
                for j in range(1, len(valid_values)):
                    if valid_values[j] == valid_values[j-1] + 1:
                        current_group.append(valid_values[j])
                    else:
                        if current_group:
                            consecutive_groups.append(current_group)
                        current_group = [valid_values[j]]
                
                if current_group:
                    consecutive_groups.append(current_group)
                
                print(f"✓ Valid untuk:", end=" ")
                group_descriptions = []
                for group in consecutive_groups:
                    if len(group) == 1:
                        group_descriptions.append(f"n = {group[0]}")
                    elif len(group) == 2:
                        group_descriptions.append(f"n = {group[0]}, {group[1]}")
                    else:
                        group_descriptions.append(f"n = {group[0]}-{group[-1]}")
                
                print(", ".join(group_descriptions))
                
                # Tambahan untuk isolated values
                isolated = [n for n in valid_values if n not in sum(consecutive_groups, [])]
                if isolated:
                    print(f"  Juga valid untuk nilai terpisah: {isolated}")
        else:
            print("✗ Tidak valid untuk range yang ditest")
        
        print()

if __name__ == "__main__":
    main()
```
